```python
import pandas as pd
from glob import glob
import numpy as np
import sys
import json
sys.path.append("..")
```

```python
lv_liste = pd.read_csv("../data/lvliste.csv")
LVs = pd.read_csv("../data/LVs.txt", sep = ";")
notenliste = pd.read_csv("../data/noteliste.csv")
studierenden = pd.read_csv("../data/studierenden.csv")


lv_information = pd.concat([lv_liste, LVs], axis=1)
lv_information = lv_information.T.drop_duplicates().T
student_data = pd.merge(left=notenliste, right=lv_information, on='lv_id', how='left')
student_data_full = pd.merge(left=student_data, right=studierenden, on = 'stud_id', how='left')
student_data_full.to_csv("../data/student_data_full.csv", index = False)
```

```python
sd = pd.read_csv("../data/student_data_full.csv")
sd_first = sd.groupby("stud_id").first()
```

```python
kat_to_lv = sd.groupby(["Kategorie", "LV Name"]).size().to_frame().drop(0, axis =1 ).reset_index()
```

```python
with open("../data/pdfinfo.json", "r") as f:
    info = json.load(f)
    
info
```

```python
lv_trans = {d["lv_name"]:d["topics"] for d in info}
kats = {1:[], 2:[], 3:[]}
for d in info:
    kats[kat_to_lv.loc[kat_to_lv["LV Name"] == d["lv_name"], "Kategorie"].values[0]].append(d["topics"])

kats
```

```python
kat_trans = {1:"Business Intelligence", 2:"Statistik und Mathematik", 3: "Informatik und Infrastruktur"}
```

```python

for kat, topic in kats.items():
    print(f"Kategorie '{kat_trans[kat]}' beinhaltet {topic}")
    
print()
```
